```python
def collect_ids(metas):
    return [meta['id'] for meta in metas]
```

```python
ids = dcard.forums('job').get_metas(num=2, callback=collect_ids)
```

```python
posts = dcard.posts(ids).get(comments=True, links=True)
```

```python
for post in posts.result():
    print(post)
    #for comment in post['comments']:
    #    print(comment)        
```

```python
for post in posts.result():
    for comment in post['comments']:
        print(comment)  
```

```python
ids = dcard.forums('job').get_metas(num=1000, callback=collect_ids)
```

```python
posts = dcard.posts(ids).get(comments=True, links=True)
```

```python
post_list = []
for post in posts.result():
    post_list.append(post)
```

```python
post_list[0]
```

```python
post_list[5]
```

```python
type(post_list[5])
```

```python
import pandas as pd
df = pd.DataFrame(columns=['id', 'school', 'forumName', 'title', 'updatedAt', 'likeCount','commentCount', 'gender', 'content'])

i=1
for post in post_list:
    df.loc[i] = [post['id'], post['school'], post['forumName'], post['title'], post['updatedAt'], post['likeCount'],\
                      post['commentCount'], post['gender'], post['content']]
    i+=1
    #print(i/len(post_list)*100)
```
